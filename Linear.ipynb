```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Data: house sizes (sq. feet) and corresponding prices
X = np.array([100, 150, 200, 250, 300]).reshape(-1, 1)  # House sizes in sq. feet
y = np.array([250000, 350000, 450000, 550000, 650000])  # Prices

# Creating the model and fitting it
model = LinearRegression()
model.fit(X, y)

# Predicting prices for new house sizes
new_house_sizes = np.array([175, 225]).reshape(-1, 1)
predicted_prices = model.predict(new_house_sizes)

# Outputting the intercept and slope of the fitted line
print("Intercept =", model.intercept_)
print("Slope =", model.coef_[0])

# Plotting the data points, regression line, and predictions
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter(new_house_sizes, predicted_prices, color='green', label='Predictions')
plt.xlabel('House Size(in sqft)')
plt.ylabel('Prices(in Lakhs)')
plt.title('Simple Example for Linear Regression')
plt.legend()
plt.show()
```
